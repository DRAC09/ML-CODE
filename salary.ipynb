```python
import pandas as pd
```

```python
data=pd.read_csv('salary_data.csv')
print(data)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
exp=data[data['YearsExperience']>2.5]
print(exp)
```

```python
slr=data[data['Salary']> 100000]
print(slr)
```

```python
minimum=data['Salary'].min()
print(minimum)
```

```python
maximum=data['Salary'].max()
print(maximum)
```

```python
avg=data['Salary'].mean()
print(avg)
```

```python
slavg=data[data['Salary']>avg]
print(slavg)
```

```python
correlation=data['YearsExperience'].corr(data['Salary'])
print(f"correlation between YearsExperience and salary: {correlation:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['YearsExperience'], data['Salary'], color='red')
plt.xlabel('YearsExperience')
plt.ylabel('salary')
plt.title('Years of Experience and Salary')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = data[['YearsExperience']]
y = data['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)

print("intercept:",model.intercept_)
print("coefficient:",model.coef_[0])
```

```python
years = [[3]]
predicted_salary = model.predict(years)
print(predicted_salary)
```

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='red')
plt.plot(data['YearsExperience'], model.predict(data[['YearsExperience']]),color = 'black')
plt.xlabel('YearsExperience')
plt.ylabel('salary')
plt.title('Years of Experience and Salary')
plt.grid(True)
plt.show()
```
