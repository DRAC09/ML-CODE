```python
import numpy as np

array1 = np.array(12)
array2 = np.array([1, 2, 3, 4, 5])
array3 = np.array([[1, 2, 3], [4, 5, 6]])
array4 = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])
array5 = np.array([1, 2, 3, 4], ndmin=5)

print('dimension of array array1:', array1.ndim)
print('dimension of array array2:', array2.ndim)
print('dimension of array array3:', array3.ndim)
print('dimension of array array4:', array4.ndim)
print('dimension of array array5:', array5.ndim)
```
